# Data processing

```python
# Let's first get all the features having percentage of missing values < 5%
cols_1 = [col for col in df.columns if df[col].isnull().mean()*100<5.0 and df[col].isnull().mean()*100>0.0]
temp1 = df[cols_1].copy()
temp1.head(5)
```

```python
# Function to find IQR , Upper_lmt and Lower_lmt
def find_upper_and_lower_lmt(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    IQR = q3-q1
    Upper_lmt = q3 + (1.5*IQR)
    Lower_lmt = q1 - (1.5*IQR)
    return [Upper_lmt,Lower_lmt]
```

```python
# Let's cap the outliers
for col in temp1.select_dtypes(include=np.float32).columns:
    lmts = find_upper_and_lower_lmt(temp1[col])
    temp1[col] = np.where(temp1[col] > lmts[0],lmts[0],
    np.where(
        temp1[col] < lmts[1],lmts[1],temp1[col] ))
```

For imputing the missing values we will use MICE algorithm, but before implementing this algorithm we need to encode the categorical feature

```python
# Let's encode the Winddir9am feature
map_wind_direction = {'W':0, 'NNW':1, 'SE':2, 'ENE':3, 'SW':4, 'SSE':5, 'S':6, 'NE':7, 'SSW':8, 'N':9, 'WSW':10,'ESE':11,'E':12,'NW':13,
                      'WNW':14, 'NNE':15}

temp3['WindDir9am'] = temp3['WindDir9am'].map(map_wind_direction)
```

```python
# Now we will be using MICE for imputing the missing values in company_size variable
lr = LinearRegression()
imp = IterativeImputer(estimator=lr, verbose=2, max_iter=30, tol=1e-10, imputation_order='roman')
```

```python
# Let's now impute the missing value
temp3 = imp.fit_transform(temp3)
```

```python
temp3 = pd.DataFrame(temp3)
temp3.columns = ['Evaporation', 'Sunshine', 'WindDir9am', 'Pressure9am', 'Pressure3pm','Cloud9am', 'Cloud3pm']
temp3 = round(temp3)
```

```python
# Now let's convert the WindDir9am feature back to categories
map_wind_direction = {0.0:'W',1.0:'NNW',2.0:'SE', 3.0:'ENE',4.0:'SW',5.0:'SSE', 6.0:'S', 7.0:'NE', 8.0:'SSW',9.0:'N',10.0:'WSW',11.0:'ESE',
                      12.0:'E',13.0:'NW',14.0:'WNW', 15.0:'NNE'}

temp3['WindDir9am'] = temp3['WindDir9am'].map(map_wind_direction)
```

```python
temp1 = df.copy()
temp1 = temp1.reset_index()
```

```python
# For all the features having sort of normal distribution let's calculate z scores
temp1['MinTemp_z'] = (temp1['MinTemp'] - temp1['MinTemp'].mean())/temp1['MinTemp'].std()
temp1['MaxTemp_z'] = (temp1['MaxTemp'] - temp1['MaxTemp'].mean())/temp1['MaxTemp'].std()
temp1['Humidity3pm_z'] = (temp1['Humidity3pm'] - temp1['Humidity3pm'].mean())/temp1['Humidity3pm'].std()
temp1['Temp9am_z'] = (temp1['Temp9am'] - temp1['Temp9am'].mean())/temp1['Temp9am'].std()
temp1['Temp3pm_z'] = (temp1['Temp3pm'] - temp1['Temp3pm'].mean())/temp1['Temp3pm'].std()
temp1['pressure9am_z'] = (temp1['pressure9am'] - temp1['pressure9am'].mean())/temp1['pressure9am'].std()
temp1['pressure3pm_z'] = (temp1['pressure3pm'] - temp1['pressure3pm'].mean())/temp1['pressure3pm'].std()
```

```python
# Let's trim the outliers
temp1 = temp1[(temp1['MinTemp_z'] > -3) | (temp1['MinTemp_z'] < 3)]
temp1 = temp1[(temp1['MaxTemp_z'] > -3) | (temp1['MaxTemp_z'] < 3)]
temp1 = temp1[(temp1['Temp9am_z'] > -3) | (temp1['Temp9am_z'] < 3)]
temp1 = temp1[(temp1['Temp3pm_z'] > -3) | (temp1['Temp3pm_z'] < 3)]
temp1 = temp1[(temp1['pressure9am_z'] > -3) | (temp1['pressure9am_z'] < 3)]
temp1 = temp1[(temp1['pressure3pm_z'] > -3) | (temp1['pressure3pm_z'] < 3)]
```

```python
# Let's check the change in distribution before and after trimming outliers
fig,ax = plt.subplots(2,3,figsize = (18,8))

sns.kdeplot(df['MaxTemp'],shade=True,color = 'green',ax = ax[0][0],label='Before')
sns.kdeplot(temp1['MaxTemp'],shade=True,color = 'red',ax = ax[0][0],label = 'After trimming')
ax[0][0].legend(loc='upper left')

sns.kdeplot(df['MinTemp'],shade=True,color = 'green',ax = ax[0][1],label='Before')
sns.kdeplot(temp1['MinTemp'],shade=True,color = 'red',ax = ax[0][1],label = 'After trimming')
ax[0][1].legend(loc='upper left')

sns.kdeplot(df['Temp9am'],shade=True,color = 'green',ax = ax[0][2],label='Before')
sns.kdeplot(temp1['Temp9am'],shade=True,color = 'red',ax = ax[0][2],label = 'After trimming')
ax[0][2].legend(loc='upper left')

sns.kdeplot(df['Temp3pm'],shade=True,color = 'green',ax = ax[1][0],label='Before')
sns.kdeplot(temp1['Temp3pm'],shade=True,color = 'red',ax = ax[1][0],label = 'After trimming')
ax[1][0].legend(loc='upper left')

sns.kdeplot(df['pressure9am'],shade=True,color = 'green',ax = ax[1][1],label='Before')
sns.kdeplot(temp1['pressure9am'],shade=True,color = 'red',ax = ax[1][1],label = 'After trimming')
ax[1][1].legend(loc='upper left')

sns.kdeplot(df['pressure3pm'],shade=True,color = 'green',ax = ax[1][2],label='Before')
sns.kdeplot(temp1['pressure3pm'],shade=True,color = 'red',ax = ax[1][2],label = 'After trimming')
ax[1][2].legend(loc='upper left')
```
